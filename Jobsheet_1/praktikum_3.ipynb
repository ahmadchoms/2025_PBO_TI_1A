```python
print("Praktikum 03. Percabangan")

# 1. IF Sederhana
# Program hanya mengeksekusi blok jika kondisinya True

nilai = 85
print("\n===== Contoh IF Sederhana =====")
print(f"Nilai Anda: {nilai}")
if nilai > 80:
    print("Selamat! Anda lulus dengan nilai tinggi. Karena nilai lebih dari 80\n")  # Pesan hanya ditampilkan jika nilai lebih dari 80


# 2. IF-ELSE
# Jika kondisi True, eksekusi blok if; jika False, eksekusi blok else

umur = 17
print("\n===== Contoh IF-ELSE =====")
print(f"Umur Anda: {umur}")
if umur >= 18:
    print("Anda sudah cukup umur untuk mendapatkan SIM. Karena umur lebih besar sama dengan 18")  # Jika umur >= 18
else:
    print("Anda belum cukup umur untuk mendapatkan SIM. Karena umur kurang dari 18\n")  # Jika umur < 18


# 3. IF-ELIF-ELSE
# Menangani banyak kondisi secara berurutan.

hari = "Rabu"
print("\n===== Contoh IF-ELIF-ELSE =====")
print(f"Hari ini adalah: {hari}")
if hari == "Senin":
    print("Hari Senin - Saatnya kembali bekerja. Jika hari bernilai Senin")  # Jika hari Senin
elif hari == "Selasa":
    print("Hari Selasa - Jadwal rapat mingguan. Jika hari bernilai Selasa")  # Jika hari Selasa
elif hari == "Rabu":
    print("Hari Rabu - Ada diskon di beberapa toko. Jika hari bernilai Rabu")  # Jika hari Rabu
else:
    print("Hari lainnya - Atur jadwalmu dengan baik. Jika hari bernilai selain dari kondisi diatas\n")  # Untuk semua hari lainnya


# 4. IF Bersarang (Nested IF)
# Kondisi di dalam kondisi untuk memeriksa sub-kondisi setelah kondisi pertama terpenuhi.

suhu = 35
print("\n===== Contoh IF Bersarang (Nested IF) =====")
print(f"Suhu saat ini: {suhu}°C")
if suhu > 30:
    print("Cuaca cukup panas. Jika suhu lebih dari 30")  # Jika suhu lebih dari 30°C
    if suhu > 40:
        print("Bahkan sangat terik! Disarankan banyak minum air. Jika suhu lebih dari 40")  # Jika suhu lebih dari 40°C
    else:
        print("Masih relatif normal, tapi tetap jaga kesehatan. Jika tidak lebih dari 40, tetapi lebih dari 30")  # Jika suhu antara 30°C dan 40°C
else:
    print("Cuaca sepertinya cukup sejuk. Jika suhu kurang dari 30\n")  # Jika suhu 30°C atau kurang


# 5. Menggabungkan Percabangan dengan Operasi Logika
# Memeriksa beberapa kondisi sekaligus dengan and, or, not

nilai_teori = 75
nilai_praktik = 80
print("\n===== Contoh IF dengan Operasi Logika AND/OR =====")
print(f"Nilai Teori: {nilai_teori}, Nilai Praktik: {nilai_praktik}")
if nilai_teori >= 70 and nilai_praktik >= 70:
    print("Anda lulus karena nilai teori dan praktik memadai. Jika nilai_teori dan nilai_praktik lebih dari sama dengan 70")  # Lulus jika kedua nilai >= 70
elif nilai_teori < 70 and nilai_praktik < 70:
    print("Anda perlu meningkatkan nilai teori dan praktik. Jika nilai_teori dan nilai_praktik kurang dari 70")  # Jika kedua nilai di bawah 70
elif nilai_teori < 70:
    print("Anda perlu meningkatkan nilai teori. Jika nilai_teori kurang dari 70")  # Jika hanya nilai teori yang kurang
else:
    print("Anda perlu meningkatkan nilai praktik. Jika nilai_praktik kurang dari 70\n")  # Jika hanya nilai praktik yang kurang


# 6. Penggunaan If Ternary (atau Conditional Expression)
# Bentuk ringkas: <hasil_if_true> if <kondisi> else <hasil_if_false>

angka = -5
print("\n===== Contoh IF Ternary (Conditional Expression) =====")
status = "Positif" if angka > 0 else "Negatif atau Nol"
print(f"Angka = {angka} => {status}")  # Menampilkan hasil positif atau negatif berdasarkan kondisi angka
```
