```python
print("Praktikum 02. Variabel dan Operasi Dasar\n")

print("Pendeklarasian Variabel tidak perlu diberi keterangan secara detail atau eksplisit")

# 1. Deklarasi Variabel
nama = "Budi"       # String
umur = 20           # Integer
tinggi = 170.5      # Float
is_student = True   # Boolean

# Menampilkan nilai variabel
print("\n===== VARIABEL =====")
print(f"Nama = {nama}, Umur = {umur}, Tinggi = {tinggi}, Mahasiswa = {is_student}  # Menampilkan variabel yang telah dideklarasikan")

# 2. Operasi Aritmatika
a = 10
b = 3

penjumlahan = a + b
pengurangan = a - b
perkalian = a * b
pembagian = a / b
modulus = a // b
pangkat = a ** b

print("\n===== OPERASI ARITMATIKA =====")
print(f"{a} + {b} = {penjumlahan}  # Operasi penjumlahan")
print(f"{a} - {b} = {pengurangan}  # Operasi pengurangan")
print(f"{a} * {b} = {perkalian}  # Operasi perkalian")
print(f"{a} / {b} = {pembagian}  # Operasi pembagian (hasil float)")
print(f"{a} // {b} = {modulus}  # Operasi modulus (hasil pembagian bulat)")
print(f"{a} ** {b} = {pangkat}  # Operasi perpangkatan")

# 3. Operasi Perbandingan
# Menghasilkan nilai Boolean (True/False)
lebih_besar = a > b
kurang_dari = a < b
sama_dengan = a == b
tidak_sama = a != b
lebih_besar_sama = a >= b
kurang_sama = a <= b

print("\n===== OPERASI PERBANDINGAN =====")
print(f"{a} > {b}  = {lebih_besar}  # Apakah a lebih besar dari b?")
print(f"{a} < {b}  = {kurang_dari}  # Apakah a lebih kecil dari b?")
print(f"{a} == {b} = {sama_dengan}  # Apakah a sama dengan b?")
print(f"{a} != {b} = {tidak_sama}  # Apakah a tidak sama dengan b?")
print(f"{a} >= {b} = {lebih_besar_sama}  # Apakah a lebih besar atau sama dengan b?")
print(f"{a} <= {b} = {kurang_sama}  # Apakah a lebih kecil atau sama dengan b?")

# 4. Operasi Logika
# Menggunakan operator logika: and, or, not
x = True
y = False

logika_and = x and y
logika_or = x or y
logika_not_x = not x

print("\n===== OPERASI LOGIKA =====")
print(f"x = {x}, y = {y}  # Menampilkan nilai boolean x dan y")
print(f"x and y = {logika_and}  # True jika keduanya True")
print(f"x or y  = {logika_or}  # True jika salah satu True")
print(f"not x   = {logika_not_x}  # Kebalikan dari nilai x")

# 5. Percabangan (if-else)
print("\n===== PERCABANGAN =====")
print("Mengecek apakah a lebih besar dari b dan b lebih besar dari 0...")
if a > b and b > 0:
    print("Kondisi terpenuhi: a lebih besar dari b, dan b masih positif.")
else:
    print("Kondisi tidak terpenuhi atau b <= 0.")
```
