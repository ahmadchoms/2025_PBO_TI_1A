```python
class AlatElektronik:
    def __init__(self, merk, model):
        self.merk = merk
        self.model = model
        self.status_daya = False

    def nyalakan(self):
        self.status_daya = True
        print(f"Alat dengan merk {self.merk} dan model {self.model} sudah dinyalakan\n")

    def matikan(self):
        self.status_daya = False
        print(f"Alat dengan merk {self.merk} dan model {self.model} sudah dimatikan\n")

    def get_info(self):
        print("Informasi Alat Elektronik")
        print(f"Merk = {self.merk}")
        print(f"Model = {self.model}")
        print(f"Status Daya = {'Hidup' if self.status_daya else 'Mati'}")


class Televisi(AlatElektronik):
    def __init__(self, merk, model, ukuran_layar):
        super().__init__(merk, model)
        self.ukuran_layar = ukuran_layar

    def get_info(self):
        super().get_info()
        print(f"Ukuran Layar = {self.ukuran_layar} Inch\n")

    def ganti_channel(self, channel):
        if self.status_daya:
            print(f"Mengganti channel ke {channel}\n")
        else:
            print("Hidupkan daya terlebih dahulu!\n")


class Radio(AlatElektronik):
    def __init__(self, merk, model, frekuensi):
        super().__init__(merk, model)
        self.frekuensi = frekuensi

    def get_info(self):
        super().get_info()
        print(f"Frekuensi = {self.frekuensi} MHz\n")

    def set_frekuensi(self, frekuensi_baru):
        if self.status_daya:
            print(f"Mengubah frekuensi ke {frekuensi_baru} MHz\n")
        else:
            print("Hidupkan daya terlebih dahulu!\n")


if __name__ == "__main__":
    televisi = Televisi('Samsung', 'Tabung', 42)
    radio = Radio('Sony', 'Pocket', 122.4)

    televisi.get_info()
    radio.get_info()

    televisi.nyalakan()
    televisi.ganti_channel(2)
    
    radio.nyalakan()
    radio.set_frekuensi(165.9)

    televisi.get_info()
    radio.get_info()

    televisi.matikan()
    radio.matikan()
```
